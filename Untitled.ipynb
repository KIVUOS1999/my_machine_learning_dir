```python
#knapsack problem for fraction distributtion

n = int(input("Enter total number of items: "))
knapsack_capacity = int(input("Enter bag capacity: "))
optimal_profit = 0

profit = []
weights = []
ppw = []
for i in range(0, n):
    profit.append(int(input("Enter profit:")))
    weights.append(int(input("Enter weights:")))
    ppw.append(profit[i]/weights[i])
    print("\n")

'''
Sorts 
profit, weights acoording to ppw(profit per unit)
ppw is sorted in decreasing order.
'''
for i in range(n):
    for j in range(n-1):
        if ppw[j]<ppw[j+1]:
            ppw[j],ppw[j+1] = ppw[j+1],ppw[j]
            profit[j],profit[j+1] = profit[j+1],profit[j]
            weights[j],weights[j+1] = weights[j+1],weights[j]

i = 0
while(knapsack_capacity>0 and i<n):
    if(weights[i]<=knapsack_capacity):
        optimal_profit+=profit[i]
        knapsack_capacity-=weights[i]
    else:
        optimal_profit+=knapsack_capacity/weights[i]*profit[i]
        knapsack_capacity=0
        print(optimal_profit)
        break
    i+=1
            
```

```python
def calculate_storage(filesize):
    block_size = 4096
    # Use floor division to calculate how many blocks are fully occupied
    full_blocks = filesize // block_size
    # Use the modulo operator to check whether there's any remainder
    partial_block_remainder = filesize%block_size
    # Depending on whether there's a remainder or not, return
    # the total number of bytes required to allocate enough blocks
    # to store your data.
    if partial_block_remainder > 0:
        return full_blocks*block_size + block_size
    return full_blocks*block_size

print(calculate_storage(1))    # Should be 4096
print(calculate_storage(4096)) # Should be 4096
print(calculate_storage(4097)) # Should be 8192
print(calculate_storage(6000)) # Should be 8192
```

```python
import pandas as pd
import numpy as np
import tqdm

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding, Flatten

csv = pd.read_csv('C:/Users/Souvik_2/Downloads/name_gender.csv', delimiter=",")
```

```python
x = []
y = []
for i in csv.T.items():
    x.append(i[1]['name'])
    y.append(i[1]['gender'])
```

```python
len(x)
```

```python
len(y)
```

```python
for i in range(len(y)):
    if y[i] == 'F':
        y[i] = 0
    else:
        y[i] = 1
```

```python
datas = 'abcdefghijklmnopqrstuvwxyz'
```

```python
token_table = {}
j = 0
for i in datas:
    token_table[i] = j/25
    j+=1
```

```python
for i in range(len(x)):
    x[i] = [token_table[i.lower()] for i in list(x[i])]
```

```python
maxx = max([len(i) for i in x])
```

```python
for i in range(len(x)):
    curr_len = len(x[i])
    padd_len = maxx - curr_len
    x[i] += [-1.0]*padd_len
```

```python
x = np.array(x)
y = np.array(y)
```

```python
X_train = x[:-500]
Y_train = y[:-500]
X_test = x[-500:]
Y_test = y[-500:]
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(4096, activation = 'relu'))
model.add(Dense(2048, activation = 'relu'))
model.add(Dense(1024, activation = 'relu'))
model.add(Dense(1024, activation = 'relu'))
model.add(Dense(526, activation = 'relu'))
model.add(Dense(526, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(2, activation = 'softmax'))

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, Y_train, epochs=20, validation_data = (X_test,Y_test))
```

```python
name = 'amit'
name = [token_table[i.lower()] for i in list(name)]
curr_len = len(name)
padd_len = maxx - curr_len
name += [-1.0]*padd_len
if np.argmax(model.predict([name])) == 0:
    print('F')
else:
    print('M')
```
